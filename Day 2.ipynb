# Visualization using Matplotlib and a gentle introduction to Machine Learning

```python
# importing all the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
time = np.arange(0, 10, 0.1)
sine = np.sin(time)
cosine = np.cos(time)
plt.title('Sine & Cosine waves', fontsize = 30)
plt.xlabel('time', fontsize = 20)
plt.ylabel('Amplitude', fontsize = 20)
plt.plot(time, cosine, color ='blue', label = 'Cosine')
plt.plot(time, sine, color = 'red', marker = 'o', label = 'Sine')
plt.legend(loc = 'lower left')
#plt.show()
```
